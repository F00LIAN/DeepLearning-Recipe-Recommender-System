### Popularity Based Recommendation System

```python
recipes['RecipeId'].astype(int)
```

```python
ratings_with_name = recipes.merge(reviews, on= 'RecipeId')
```

```python
num_rating_df = ratings_with_name.groupby('Name').count()['Review'].reset_index()
num_rating_df.rename(columns = {'Review': 'num_of_ratings'}, inplace = True)
num_rating_df
```

```python
ratings_with_name
```

```python
avg_rating_df = ratings_with_name.groupby('Name').mean()['Rating'].reset_index()
avg_rating_df.rename(columns = {'Rating': 'avg_rating'}, inplace = True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on = 'Name')
popular_df
```

```python
popular_df['num_of_ratings'].describe()
```

```python
#Top 50 Recipes with 50 or more ratings from users
popular_df = popular_df[popular_df['num_of_ratings'] >= 50].sort_values('avg_rating', ascending=False)
popular_df = popular_df.merge(recipes, on='Name').drop_duplicates('Name')[['RecipeId', 'Name', 'RecipeIngredientParts', 'RecipeCategory', 'Calories', 'ProteinContent', 'CarbohydrateContent', 'Description', 'RecipeInstructions', 'num_of_ratings', 'avg_rating']]
```

### Collaborative Filtering Based Recommender System based off Users

```python
# find users that have rated over 200 recipes = x
x = ratings_with_name.groupby('AuthorId_x').count()['Rating'] > 200
trusted_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['AuthorId_x'].isin(trusted_users)]
```

```python
# Recipes with the most ratings by users = y
y = filtered_rating.groupby('Name').count()['Rating'] >= 50
famous_recipes = y[y].index
```

```python
# Create a DataFrame containing the most famous_recipes rated by the most trusted users
final_ratings = filtered_rating[filtered_rating['Name'].isin(famous_recipes)]
```

```python
pt = final_ratings.pivot_table(index='Name', columns = 'AuthorId_x', values = 'Rating')
pt.fillna(0, inplace=True)
```

```python
# Trusted Reviewers are columns
# Recipes are rows
pt
```

```python
#Find Cosine Similarity for matrix above
similarity_scores = cosine_similarity(pt)
similarity_scores
```

```python
# Create recommendation function
def recommend(recipe_name):
    # index fetch
    index = np.where(pt.index == recipe_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])), key = lambda x:x[1], reverse = True)[1:6]
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend('Zucchini Pancakes')
```

```python
recommend('garlic shrimp with asparagus')
```

```python
recommend('Banana Coconut Muffins')
```

```python
recommend('Indian Butter Chicken')
```

## Recommender System Part 3 
### Build a recommendation system using Embeddings from the HuggingFace Library 
### Transformers are widely used today for NLP applications. They are trained on massive amounts of data and are now accessible for everyone for downstream small data tasks. 
### Performs much better than Tfidf and Cosine, since it captures context better. With this we can input food category, type, or ingredients to help us get food recommendations
### I included a visualization using PCA to allow a 2D representation of how the nearest neighbors algorithm will find distances. 
### We can also try and impliment a chatbot or further improve the recommendations by returning ingredient lists aswell. 


```python
#! pip install sentence-transformers
```

```python
from transformers import AutoTokenizer, AutoModel
from sentence_transformers import SentenceTransformer
from sklearn.neighbors import NearestNeighbors
from sklearn.decomposition import PCA
```

```python
df1 = popular_df
```

```python
df1.shape
```

```python
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

# Convert pandas Series to list of strings
description = df1['Description'].dropna().tolist()

# Generate embeddings for list of strings
embeddings = model.encode(description)

def embed(texts):
    return model.encode(texts)

pca = PCA(n_components=2)
emb_viz = pca.fit_transform(embeddings)

nn = NearestNeighbors(n_neighbors=10)
nn.fit(embeddings)

def recommend(text):
    emb = embed([text])
    neighbors = nn.kneighbors(emb, return_distance=False)[0]
    return df1[['Name', 'RecipeIngredientParts']].iloc[neighbors]#.tolist()

print('Recipes')
recommend('Mexican')
```

```python
#PCA visualization of embedding space 
plt.title('Embedding Space')
plt.scatter(emb_viz[:, 0], emb_viz[:, 1])
plt.show()
```
